```python
class Processor:
    def __init__(self):
        self.arabic_diacritics = re.compile("""
                                         ّ    | # Tashdid
                                         َ    | # Fatha
                                         ً    | # Tanwin Fath
                                         ُ    | # Damma
                                         ٌ    | # Tanwin Damm
                                         ِ    | # Kasra
                                         ٍ    | # Tanwin Kasr
                                         ْ    | # Sukun
                                         ـ     # Tatwil/Kashida
                                     """, re.VERBOSE)

    def normalize_arabic(self, text):
        text = re.sub("[إأآاٱ]", "ا", text)
        text = re.sub("ة", "ه", text)
        text = re.sub("گ", "ك", text)
        text = re.sub(r'\sو(\w+)', r' و \1', text)
        return text

    def stem_and_process(self, text):
        processed_text = self.arabic_diacritics.sub('',text)
        return processed_text

# Create an instance of the text processor
text_processor = Processor()
```

```python
df["cases_text"] = df['judgment_regex'].apply(lambda x: text_processor.stem_and_process(x))
df['cases_text'] = df['cases_text'].apply(lambda x: ' '.join(x.split()))
```
